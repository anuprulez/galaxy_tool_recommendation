```python
dist_freq = "/home/kumara/Desktop/dist_freq.json"
```

```python
import json

with open(dist_freq, "r") as file:
    freq = json.loads(file.read())
```

```python
sorted_freq = dict(sorted(freq.items(), key=lambda kv: kv[1], reverse=True))
```

```python
sorted_freq
```

```python
max_freq = max(list(sorted_freq.values()))
```

```python
inv_freq = dict()
for k in sorted_freq:
    inv_freq[k] = int(max_freq / float(sorted_freq[k]))
```

```python
inv_freq
```
