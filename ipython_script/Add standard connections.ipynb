```python
import numpy as np
import json
import warnings
import operator

import h5py
from keras.models import model_from_json
from keras import backend as K

warnings.filterwarnings("ignore")
```

```python
model_file_path = "data/tool_recommendation_model_original.hdf5"
standard_connections_path = "data/standard_connections.json"
with open(standard_connections_path, "r") as sc_file:
    standard_connections = json.loads(sc_file.read())
#print(standard_connections)
trained_model = h5py.File(model_file_path, 'a')
try:
    trained_model.create_dataset("standard_connections", data=json.dumps(standard_connections))
except:
    trained_model.modify("standard_connections", data=json.dumps(standard_connections))
trained_model.close()
```
