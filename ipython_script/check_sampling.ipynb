```python
import json
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

def get_freq(sampled_t_freq):
    with open(sampled_t_freq, "r") as f:
        tool_freq = json.loads(f.read())
    return tool_freq

def order_tools(s_t, o_t):
    s_list = list()
    o_list = list()
    for t in s_t:
        o_f = o_t[t]
        s_f = s_t[t]
        o_list.append(o_f)
        s_list.append(s_f)
    return s_list, o_list

def plot_freq(tool_freq, plt_title):
    fig_size = (9, 9)
    x = [i for i in range(len(tool_freq))]
    plt.figure(figsize=fig_size)
    plt.bar(x, tool_freq)
    plt.title(plt_title)
    plt.grid(True)
    plt.show()

base_p = "../output_files/data_20_05/gru_wc/run1/"
orig_t_freq = get_freq(base_p + "freq_dict_names.txt")
sampled_t_freq = get_freq(base_p + "generated_tool_frequencies.txt")

s_freq, o_freq = order_tools(sampled_t_freq, orig_t_freq)

plot_freq(o_freq, "Original frequency of tools")


plot_freq(s_freq, "Sampled frequency of tools after one epoch")
```
